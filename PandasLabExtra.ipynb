For this lab, you will need to read in the provided FASTA file and generate a DataFrame containing the following information for each sequence record:
1. Sequence Length
2. GC content
3. Begins with start codon?
4. Ends in stop codon?
5. Complete frame? (Meaning is the length a multiple of 3?)

The row index should be the sequence label, and columns should be labeled with descriptions of these 5 criteria I've given you. Be mindful of the datatype you choose for description 3-5.

```python
import pandas as pd
columnList=["Sequence Length", "GC%", "Valid Start", "Valid Stop", "Complete Frame"]
```

To read the fasta file and generate the header and sequence to create a dictionary of it

```python
def readMultilineFASTA(file):
    name = None
    seq = ""
    for line in file:
        if line.startswith(">"): # To differentiate between header and sequences
            if name:
                yield (name,seq)
                name = None
                seq = ""
            name = line.rstrip().lstrip(">")
        else:
            seq += line.rstrip()

    yield name,seq # yield the name(header) and seq to create dictionary
```

Code to read the dictionary of header and sequence and generate a new dictionary with header as key and value as the list of 5 criterias

```python
def createNewDict(dictionary):
    newDict = {}
    SequenceLength = 0
    GCper = 0, 
    Valid_Start = False
    startCodon = ["AUG", "GUG", "UUG"] # start codon list
    Valid_Stop = False
    endCodon = ["UAG","UAA","UGA"] # end codon list
    CompleteFrame = False

    for k,v in dictionary.items(): # for all the key and value in the dictionary
        criteriaList = [] # temp list to hold the 5 criterias
        SequenceLength = len(v) # length of the sequence
        criteriaList.append(SequenceLength) # append to the temp list the length of the sequence
        
        GCper = (v.count("G") + v.count("C"))/SequenceLength * 100 # GC percent of the sequence
        criteriaList.append(int(GCper)) # append to the temp list GC percent of the sequence
        
        for startCodon in startCodon: # for every codon in the start codon list, we will see if the codon matches the starting of the sequence 
            if v[:len(startCodon)] == startCodon:
                Valid_Start = True # if the first elements matches then make the valid start as true
        criteriaList.append(Valid_Start) # append to the value of the valid start either true of false
        
        for endCodon in endCodon: # for every codon in the stop codon list, we will see if the codon matches the end of the sequence
            if v[SequenceLength-len(endCodon):] == endCodon:
                Valid_Stop = True # if the last elements matches then make the valid start as true
        criteriaList.append(Valid_Stop) # append to the value of the valid stop either true of false
        
        if SequenceLength%3 == 0: # if the length of the sequence is divisible by 3 then CompleteFrame is true
            CompleteFrame = True
        criteriaList.append(CompleteFrame) # append to the value of the CompleteFrame either true of false
        
        newDict[k] = criteriaList # create a new dictionary with the key as header and value as the temp list containing the 5 criteria of the header sequence
        
    return newDict # return the new Dictionary 
```

Main function to read the fasta file and pass on to the function for further steps

```python
def main():
    d = {}

    file = r"Mdomestica-pandasLab.fa"
    f = open(file, 'r')
    data = f.readlines()
    count = 0
    for headSeq in readMultilineFASTA(data):
        d[headSeq[0][:12]] = headSeq[1] # create a dictionary after reading the file with header as key and sequence as value
    
    return createNewDict(d)
```

```python
newDict = main() # the new dictionary with the key as header and value as the temp list containing the 5 criteria of the header sequence
df = pd.DataFrame(newDict, index=columnList) # create a dataframe from the new dictionary using the column list
```
